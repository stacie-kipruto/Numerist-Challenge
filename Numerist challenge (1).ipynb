```python
x = landing_plot['Month of Year']
y = landing_plot['Pageviews']

plt.figure(figsize = (10,8), dpi = 90)
plt.plot(x,y, label = 'Month of Year', color = 'red')
plt.xlabel('Month of Year')
plt.ylabel('Pageviews')
plt.title('New users over months for 25 rows')
plt.legend()
plt.show()
```
